### Function how to define and use

```python
def remove_item(input_list,item_to_remove):
    """Removes item_to_remove from input_list"""
    
    output_list = []
    for element in input_list:
        if element == item_to_remove:
            pass
        else:
            output_list.append(element)
          
    return output_list
```

```python
my_list = [1,2,3,4,5,6,7,8]
result = remove_item(my_list,5)
print(my_list)
print(result)
```

```python
my_list = [1,2,3,4,5,6,7,8]
result = my_list.remove(5)
print(my_list)
print(result)
```

What if we want to stop adding numbers? Adding a condition

```python
values = []

for i in range(100):
    this_value = input()
    if this_value == 'exit':
        break    
    this_value = int(this_value)
    values.append(this_value)
```

```python
values
```

```python
values = []
n_sessions = 2
flag = 0

for session in range(n_sessions):
    session_values = []
    print(f'You are in session {session}')
    for i in range(1000000000):
        this_value = input()
        if this_value == 'exit':          
            break    
        this_value = int(this_value)
        session_values.append(this_value)
        
    # save to the big overall list
    values.append(session_values)
```
