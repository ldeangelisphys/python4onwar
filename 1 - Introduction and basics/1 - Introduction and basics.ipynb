Now we make sure that values is a list of numbers

```python
counter = []
values = []

for i in range(5):
    counter.append(i)
    this_value = input()
    this_value = int(this_value)
    values.append(this_value)
```

What if we want to stop adding numbers? Adding a condition

```python
values = []

for i in range(100):
    this_value = input()
    if this_value == 'exit':
        break    
    this_value = int(this_value)
    values.append(this_value)
```

```python
values
```

```python
values = []
flag = 0
n_sessions = 2


for session in range(n_sessions):
    session_values = []
    print('You are in session {}'.format(session))
    for i in range(100):
        this_value = input()
        if this_value == 'exit':
            flag = 1
            break    
        this_value = int(this_value)
        session_values.append(this_value)
        
    # save to the big overall list
    values.append(session_values)
```
