# Functions

Standard function definition

```python
def str2int(s):
    
    return int(s)



def my_function(par1,par2,par3=1):
    
    par1 = str2int(par1)
    par2 = str2int(par2)
    
    product = par1*par2*par3
    
    return product


def list_product(A):
    """Computes the product of the elements in a list
    
    A : list of parameters to be multiplied   
    """
    
    prod = 1
    
    for el in A:
        prod = prod * el
        # prod *= el Exactly equivalent above
        
    return prod
```
